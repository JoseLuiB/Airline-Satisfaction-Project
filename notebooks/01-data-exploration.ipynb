#### Recolección de Datos
Los datos utilizados en este proyecto fueron extraídos de Kaggle (https://www.kaggle.com/datasets/teejmahal20/airline-passenger-satisfaction). Estos datos consisten en dos archivos CSV que contienen información sobre la satisfacción de los clientes de una aerolínea. Los archivos han sido descargados y guardados en la carpeta `data/raw/`.

## *Carga y Exploracón de Datos*

```python
# Importamos librerias
import pandas as pd
```

```python
# Cargamos los datos y los unimos en un unico data frame
data1 = pd.read_csv("../data/raw/airline_satisfaction_part1.csv")
data2 = pd.read_csv("../data/raw/airline_satisfaction_part2.csv")

# Creamos un dataframe unificado
df = pd.concat([data1, data2], ignore_index=True)
```

```python
# Visualizas las primeras filas del DataFrame, lo cual es útil para obtener una vista inicial de los datos.
df.head()
```

```python
# Visualizas las últimas filas del DataFrame
df.tail()
```

```python
# Revisamos la estructura del DataFrame, tipos de datos y la cantidad de valores no nulos.
df.info()
```

```python
# Generamos estadísticas descriptivas de las variables numéricas, lo cual es útil para entender la distribución de los datos.
df.describe()
```

### Identificación de Problemas Iniciales

```python
# Validamos la cantidad de valores nulos por columna
df.isnull().sum()
```

```python
# Validamo la cantidad de filas duplicadas
df.duplicated().sum()
```

```python
# Filtrar las columnas no numéricas
colums_category = df.select_dtypes(exclude=['number']).columns

# Crear el ciclo for para contar los valores de cada columna categórica
for col in colums_category:
    conteo = df[col].value_counts()
    print(f"Conteo para la columna '{col}':")
    print(conteo)
    print()
```

### Guardado del dataframe combinado

```python
df.to_csv("../data/processed/combined_data.csv", index=False) 
```
