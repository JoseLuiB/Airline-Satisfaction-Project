```python
#1-Importación de librerías
import pandas as pd
import numpy as nu
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga de datos y combinación de los mismos
#Carga de datos
data1 = pd.read_csv("../data/airline_satisfaction_part1.csv")
data2 = pd.read_csv("../data/airline_satisfaction_part2.csv")

#Combinar los dataframes
data = pd.concat([data1, data2], ignore_index=True)

#Mostrar primeras filas del dataframe combinado
data.head()
```

```python
#Inspeccionar la estructura y el contenido de los datos
#Información general del dataframe
data.info()

#Estadísticas descriptivas
data.describe()
```

```python
#Histograma de las variables numéricas
data.hist(figsize=(25, 15), bins=30)
plt.show()
```
